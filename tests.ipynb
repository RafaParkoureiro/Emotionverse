```python
# Define target variable - only sentiment
y_sentiment = labels_df['sentiment']

# Split data for each feature type
X_tfidf_train, X_tfidf_test, y_train_tfidf, y_test_tfidf = train_test_split(
    X_tfidf, y_sentiment, test_size=0.2, random_state=420, stratify=y_sentiment
)

X_bow_train, X_bow_test, y_train_bow, y_test_bow = train_test_split(
    X_bow, y_sentiment, test_size=0.2, random_state=420, stratify=y_sentiment
)

X_emb_train, X_emb_test, y_train_emb, y_test_emb = train_test_split(
    X_embeddings, y_sentiment, test_size=0.2, random_state=420, stratify=y_sentiment
)

X_tfidf_emb_train, X_tfidf_emb_test, y_train_tfidf_emb, y_test_tfidf_emb = train_test_split(
    X_tfidf_embeddings, y_sentiment, test_size=0.2, random_state=420, stratify=y_sentiment
)

# Define models for sparse matrix features (TF-IDF and BoW)
sparse_models = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0),
    'Linear SVM': LinearSVC(C=1.0, max_iter=10000)
}

# Define models for dense matrix features (embeddings)
dense_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'Linear SVM': LinearSVC(C=1.0, max_iter=10000)
}

# Function to evaluate models and display results
def evaluate_models(X_train, X_test, y_train, y_test, models, feature_name):
    results = {}
    print(f"\n----- Sentiment Classification Results ({feature_name}) -----")
    
    for name, model in models.items():
        print(f"\nTraining {name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Store model performance
        report = classification_report(y_test, y_pred, output_dict=True)
        results[name] = {
            'accuracy': report['accuracy'],
            'weighted_f1': report['weighted avg']['f1-score'],
            'model': model
        }
        
        # Print classification report
        print(classification_report(y_test, y_pred))
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=sorted(y_sentiment.unique()),
                   yticklabels=sorted(y_sentiment.unique()))
        plt.title(f'Confusion Matrix - {name} ({feature_name})')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig(f'confusion_matrix_{feature_name}_{name.replace(" ", "_")}.png')
        plt.show()
        
        # Show top misclassifications
        wrong_indices = np.where(y_test != y_pred)[0]
        if len(wrong_indices) > 0:
            print("\nTop misclassifications:")
            for idx in wrong_indices[:5]:  # Show first 5 misclassifications
                print(f"True: {y_test.iloc[idx]}, Predicted: {y_pred[idx]}")
    
    return results

# Evaluate all feature types
tfidf_results = evaluate_models(X_tfidf_train, X_tfidf_test, y_train_tfidf, y_test_tfidf, sparse_models, "TF-IDF")
bow_results = evaluate_models(X_bow_train, X_bow_test, y_train_bow, y_test_bow, sparse_models, "Bag of Words")
emb_results = evaluate_models(X_emb_train, X_emb_test, y_train_emb, y_test_emb, dense_models, "Embeddings")
tfidf_emb_results = evaluate_models(X_tfidf_emb_train, X_tfidf_emb_test, y_train_tfidf_emb, y_test_tfidf_emb, dense_models, "TF-IDF Weighted Embeddings")

# Combine all results for comparison
all_results = {}
for feature_name, results in zip(
    ["TF-IDF", "Bag of Words", "Embeddings", "TF-IDF Weighted Embeddings"],
    [tfidf_results, bow_results, emb_results, tfidf_emb_results]
):
    for model_name, metrics in results.items():
        key = f"{feature_name} - {model_name}"
        all_results[key] = metrics

# Create comparison table
comparison_df = pd.DataFrame({
    'Model': [k for k in all_results.keys()],
    'Accuracy': [v['accuracy'] for v in all_results.values()],
    'F1 Score (weighted)': [v['weighted_f1'] for v in all_results.values()]
})

# Sort by accuracy
comparison_df = comparison_df.sort_values('Accuracy', ascending=False)

# Display and save comparison results
print("\n----- Overall Model Comparison -----")
print(comparison_df)
comparison_df.to_csv('sentiment_model_comparison.csv', index=False)

# Plot comparison bar chart
plt.figure(figsize=(12, 8))
comparison_df.plot(x='Model', y=['Accuracy', 'F1 Score (weighted)'], kind='bar', figsize=(12, 8))
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('sentiment_model_comparison.png')
plt.show()

# Find the best model
best_model_row = comparison_df.iloc[0]
best_model_name = best_model_row['Model']
print(f"\nBest performing model: {best_model_name}")
print(f"Accuracy: {best_model_row['Accuracy']:.4f}")
print(f"F1 Score: {best_model_row['F1 Score (weighted)']:.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import load_npz, csr_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load your features and labels
features_dir = 'features'
X_tfidf = load_npz(os.path.join(features_dir, 'tfidf_features.npz'))
labels_df = pd.read_csv(os.path.join(features_dir, 'emotionverse_labels.csv'))

# Define target variable - only sentiment
y_sentiment = labels_df['sentiment']

# First, split into train and test sets (80/20)
X_train_initial, X_test, y_train_initial, y_test = train_test_split(
    X_tfidf, y_sentiment, test_size=0.2, random_state=42, stratify=y_sentiment
)

# For the original approach, we'll use this initial split
X_train_original = X_train_initial
y_train_original = y_train_initial

# Check class distribution in original training set
print("\nClass distribution in original training set:")
print(Counter(y_train_original))

# For the SMOTE approach, we'll apply SMOTE to the training data
print("\nApplying SMOTE to balance classes...")
smote = SMOTE(random_state=42)
X_train_dense = X_train_initial.toarray()  # Convert sparse to dense
X_train_smote, y_train_smote = smote.fit_resample(X_train_dense, y_train_initial)

# Check class distribution after SMOTE
print("\nClass distribution after SMOTE:")
print(Counter(y_train_smote))

# Plot class distribution before and after SMOTE
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(y=y_train_original)
plt.title('Class Distribution Before SMOTE')
plt.xlabel('Count')
plt.ylabel('Sentiment Class')

plt.subplot(1, 2, 2)
sns.countplot(y=y_train_smote)
plt.title('Class Distribution After SMOTE')
plt.xlabel('Count')
plt.ylabel('Sentiment Class')

plt.tight_layout()
plt.savefig('sentiment_class_distribution_smote.png')
plt.show()

# Define models
models = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0),
    'Linear SVM': LinearSVC(C=1.0, max_iter=10000)
}

# Function to evaluate models and display results
def evaluate_models(X_train, X_test, y_train, y_test, models, is_smote=False):
    results = {}
    suffix = "with SMOTE" if is_smote else "without SMOTE"
    print(f"\n----- Sentiment Classification Results (TF-IDF {suffix}) -----")
    
    for name, model in models.items():
        print(f"\nTraining {name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Store model performance
        report = classification_report(y_test, y_pred, output_dict=True)
        results[name] = {
            'accuracy': report['accuracy'],
            'weighted_f1': report['weighted avg']['f1-score'],
            'model': model
        }
        
        # Print classification report
        print(classification_report(y_test, y_pred))
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=sorted(y_sentiment.unique()),
                   yticklabels=sorted(y_sentiment.unique()))
        plt.title(f'Confusion Matrix - {name} (TF-IDF {suffix})')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig(f'confusion_matrix_tfidf_{suffix.replace(" ", "_")}_{name.replace(" ", "_")}.png')
        plt.show()
        
        # Show top misclassifications
        wrong_indices = np.where(y_test != y_pred)[0]
        if len(wrong_indices) > 0:
            print("\nTop misclassifications:")
            for idx in wrong_indices[:5]:  # Show first 5 misclassifications
                try:
                    print(f"True: {y_test.iloc[idx]}, Predicted: {y_pred[idx]}")
                except AttributeError:
                    # If y_test is not a pandas Series with iloc
                    print(f"True: {y_test[idx]}, Predicted: {y_pred[idx]}")
    
    return results

# Evaluate models without SMOTE (original data)
# Make sure X_train_original is sparse for Naive Bayes
X_train_original_sparse = csr_matrix(X_train_original)
original_results = evaluate_models(X_train_original_sparse, X_test, y_train_original, y_test, models, is_smote=False)

# Evaluate models with SMOTE
# Convert SMOTE-generated data back to sparse format
X_train_smote_sparse = csr_matrix(X_train_smote)
smote_models = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0),
    'Linear SVM': LinearSVC(C=1.0, max_iter=10000)
}
smote_results = evaluate_models(X_train_smote_sparse, X_test, y_train_smote, y_test, smote_models, is_smote=True)

# Create comparison table
rows = []

# Add original results
for name, metrics in original_results.items():
    rows.append({
        'Model': name,
        'Training Method': 'Without SMOTE',
        'Accuracy': metrics['accuracy'],
        'F1 Score (weighted)': metrics['weighted_f1']
    })

# Add SMOTE results
for name, metrics in smote_results.items():
    rows.append({
        'Model': name,
        'Training Method': 'With SMOTE',
        'Accuracy': metrics['accuracy'],
        'F1 Score (weighted)': metrics['weighted_f1']
    })

# Create DataFrame from rows
comparison_df = pd.DataFrame(rows)

# Sort by accuracy
comparison_df = comparison_df.sort_values('Accuracy', ascending=False)

# Display and save comparison results
print("\n----- Overall Model Comparison -----")
print(comparison_df)
comparison_df.to_csv('sentiment_model_comparison_smote.csv', index=False)

# Plot comparison bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Model', y='Accuracy', hue='Training Method', data=comparison_df)
plt.title('Model Performance Comparison')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('sentiment_model_comparison_smote.png')
plt.show()

# Find the best model
best_model_row = comparison_df.iloc[0]
best_model_name = best_model_row['Model']
best_training = best_model_row['Training Method']
print(f"\nBest performing model: {best_model_name} ({best_training})")
print(f"Accuracy: {best_model_row['Accuracy']:.4f}")
print(f"F1 Score: {best_model_row['F1 Score (weighted)']:.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import load_npz, csr_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load your features and labels
features_dir = 'features'
X_tfidf = load_npz(os.path.join(features_dir, 'tfidf_features.npz'))
labels_df = pd.read_csv(os.path.join(features_dir, 'emotionverse_labels.csv'))

# Define target variable - only sentiment
y_sentiment = labels_df['sentiment']

# First, split into train and test sets (80/20)
X_train_initial, X_test_initial, y_train_initial, y_test_initial = train_test_split(
    X_tfidf, y_sentiment, test_size=0.2, random_state=42, stratify=y_sentiment
)

print("\n--- Initial datasets ---")
print("Training set class distribution:")
print(Counter(y_train_initial))
print("\nTest set class distribution:")
print(Counter(y_test_initial))

# Convert to dense format for SMOTE
X_train_dense = X_train_initial.toarray()
X_test_dense = X_test_initial.toarray()

# Apply SMOTE to both training and test sets
print("\n--- Applying SMOTE to both training and test sets ---")
smote = SMOTE(random_state=42)

# Apply SMOTE to training data
X_train_smote, y_train_smote = smote.fit_resample(X_train_dense, y_train_initial)
print("\nTraining set class distribution after SMOTE:")
print(Counter(y_train_smote))

# Apply SMOTE to test data
X_test_smote, y_test_smote = smote.fit_resample(X_test_dense, y_test_initial)
print("\nTest set class distribution after SMOTE:")
print(Counter(y_test_smote))

# Plot class distributions before and after SMOTE
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Before SMOTE
sns.countplot(y=y_train_initial, ax=axs[0, 0])
axs[0, 0].set_title('Training Set Before SMOTE')
axs[0, 0].set_xlabel('Count')
axs[0, 0].set_ylabel('Sentiment Class')

sns.countplot(y=y_test_initial, ax=axs[0, 1])
axs[0, 1].set_title('Test Set Before SMOTE')
axs[0, 1].set_xlabel('Count')
axs[0, 1].set_ylabel('Sentiment Class')

# After SMOTE
sns.countplot(y=y_train_smote, ax=axs[1, 0])
axs[1, 0].set_title('Training Set After SMOTE')
axs[1, 0].set_xlabel('Count')
axs[1, 0].set_ylabel('Sentiment Class')

sns.countplot(y=y_test_smote, ax=axs[1, 1])
axs[1, 1].set_title('Test Set After SMOTE')
axs[1, 1].set_xlabel('Count')
axs[1, 1].set_ylabel('Sentiment Class')

plt.tight_layout()
plt.savefig('class_distributions_smote.png')
plt.show()

# Define models
models = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=1000, C=1.0),
    'Linear SVM': LinearSVC(C=1.0, max_iter=10000)
}

# Function to evaluate models and display results
def evaluate_model(name, model, X_train, y_train, X_test, y_test, test_type):
    print(f"\n--- Training {name} on {test_type} data ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    report = classification_report(y_test, y_pred, output_dict=True)
    accuracy = report['accuracy']
    weighted_f1 = report['weighted avg']['f1-score']
    
    # Print classification report
    print(f"Classification Report for {name} on {test_type} data:")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=sorted(np.unique(y_test)),
               yticklabels=sorted(np.unique(y_test)))
    plt.title(f'Confusion Matrix - {name} ({test_type})')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f'confusion_matrix_{name.replace(" ", "_")}_{test_type}.png')
    plt.show()
    
    return {
        'model': name,
        'scenario': test_type,
        'accuracy': accuracy,
        'weighted_f1': weighted_f1
    }

# Run different training and testing scenarios
results = []

# Convert back to sparse matrices for models that prefer them
X_train_initial_sparse = csr_matrix(X_train_initial)
X_test_initial_sparse = csr_matrix(X_test_initial)
X_train_smote_sparse = csr_matrix(X_train_smote)
X_test_smote_sparse = csr_matrix(X_test_smote)

# Scenario 1: Original training data â†’ Original test data
print("\n=== SCENARIO 1: Original training data â†’ Original test data ===")
for name, model in models.items():
    result = evaluate_model(name, model, X_train_initial_sparse, y_train_initial, 
                            X_test_initial_sparse, y_test_initial, "Originalâ†’Original")
    results.append(result)

# Scenario 2: SMOTE training data â†’ Original test data
print("\n=== SCENARIO 2: SMOTE training data â†’ Original test data ===")
for name, model in models.items():
    result = evaluate_model(name, model, X_train_smote_sparse, y_train_smote, 
                            X_test_initial_sparse, y_test_initial, "SMOTEâ†’Original")
    results.append(result)

# Scenario 3: Original training data â†’ SMOTE test data
print("\n=== SCENARIO 3: Original training data â†’ SMOTE test data ===")
for name, model in models.items():
    result = evaluate_model(name, model, X_train_initial_sparse, y_train_initial, 
                            X_test_smote_sparse, y_test_smote, "Originalâ†’SMOTE")
    results.append(result)

# Scenario 4: SMOTE training data â†’ SMOTE test data
print("\n=== SCENARIO 4: SMOTE training data â†’ SMOTE test data ===")
for name, model in models.items():
    result = evaluate_model(name, model, X_train_smote_sparse, y_train_smote, 
                            X_test_smote_sparse, y_test_smote, "SMOTEâ†’SMOTE")
    results.append(result)

# Create comparison DataFrame
comparison_df = pd.DataFrame(results)

# Sort by accuracy
comparison_df = comparison_df.sort_values('accuracy', ascending=False)

# Display and save comparison results
print("\n----- Overall Model Comparison -----")
print(comparison_df)
comparison_df.to_csv('sentiment_model_comparison_all_scenarios.csv', index=False)

# Plot comparison of all scenarios
plt.figure(figsize=(12, 8))
sns.barplot(x='model', y='accuracy', hue='scenario', data=comparison_df)
plt.title('Model Performance Comparison Across Scenarios')
plt.ylabel('Accuracy')
plt.xlabel('Model')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('model_comparison_all_scenarios.png')
plt.show()

# Find the best model
best_model_row = comparison_df.iloc[0]
print(f"\nBest performing model: {best_model_row['model']} ({best_model_row['scenario']})")
print(f"Accuracy: {best_model_row['accuracy']:.4f}")
print(f"F1 Score: {best_model_row['weighted_f1']:.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import load_npz, csr_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load your features and labels
features_dir = 'features'
X_tfidf = load_npz(os.path.join(features_dir, 'tfidf_features.npz'))
labels_df = pd.read_csv(os.path.join(features_dir, 'emotionverse_labels.csv'))
y_sentiment = labels_df['sentiment']

# Split data into train and test sets
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(
    X_tfidf, y_sentiment, test_size=0.2, random_state=42, stratify=y_sentiment
)

print("Original data class distribution:")
print("Training:", Counter(y_train_orig))
print("Testing:", Counter(y_test_orig))

# Apply SMOTE to training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_orig.toarray(), y_train_orig)
print("\nAfter SMOTE - Training data class distribution:")
print(Counter(y_train_smote))

# Apply SMOTE to test data for scenario 3
X_test_smote, y_test_smote = smote.fit_resample(X_test_orig.toarray(), y_test_orig)
print("After SMOTE - Test data class distribution:")
print(Counter(y_test_smote))

# Convert to sparse matrices for SVM
X_train_orig_sparse = csr_matrix(X_train_orig)
X_test_orig_sparse = csr_matrix(X_test_orig)
X_train_smote_sparse = csr_matrix(X_train_smote)
X_test_smote_sparse = csr_matrix(X_test_smote)

# Function to evaluate LinearSVC and display results
def evaluate_svm(X_train, y_train, X_test, y_test, scenario_name):
    print(f"\n--- Scenario: {scenario_name} ---")
    
    # Train Linear SVM
    svm = LinearSVC(C=1.0, max_iter=10000, random_state=42)
    svm.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = svm.predict(X_test)
    
    # Print classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=sorted(np.unique(y_test)),
               yticklabels=sorted(np.unique(y_test)))
    plt.title(f'Confusion Matrix - LinearSVC ({scenario_name})')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f'confusion_matrix_svm_{scenario_name.replace(" ", "_")}.png')
    plt.show()
    
    return {
        'scenario': scenario_name,
        'accuracy': np.mean(y_pred == y_test),
        'model': svm
    }

# Run the three scenarios
results = []

# Scenario 1: Original â†’ Original
results.append(evaluate_svm(
    X_train_orig_sparse, y_train_orig, 
    X_test_orig_sparse, y_test_orig, 
    "Original â†’ Original"
))

# Scenario 2: SMOTE â†’ Original
results.append(evaluate_svm(
    X_train_smote_sparse, y_train_smote, 
    X_test_orig_sparse, y_test_orig, 
    "SMOTE â†’ Original"
))

# Scenario 3: SMOTE â†’ SMOTE
results.append(evaluate_svm(
    X_train_smote_sparse, y_train_smote, 
    X_test_smote_sparse, y_test_smote, 
    "SMOTE â†’ SMOTE"
))

# Create comparison DataFrame
comparison_df = pd.DataFrame(results)
print("\n--- Results Summary ---")
print(comparison_df[['scenario', 'accuracy']])

# Plot comparison of all scenarios
plt.figure(figsize=(10, 6))
sns.barplot(x='scenario', y='accuracy', data=comparison_df)
plt.title('LinearSVC Performance Comparison Across Scenarios')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('svm_comparison_scenarios.png')
plt.show()

# Find the best scenario
best_scenario = comparison_df.loc[comparison_df['accuracy'].idxmax()]
print(f"\nBest scenario: {best_scenario['scenario']} with accuracy: {best_scenario['accuracy']:.4f}")
```
