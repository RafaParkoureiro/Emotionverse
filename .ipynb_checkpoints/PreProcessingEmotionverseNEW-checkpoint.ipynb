## PreProcessing

```python
# Preprocessing functions
def clean_text(text):
    text = re.sub(r'http\S+|www\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove @mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)  # Remove emojis (Unicode range)
    return text

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def to_lowercase(text):
    return text.lower()

def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])

def lemmatize_text(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
# Preprocessing pipeline
def preprocess_text_pipeline(text):
    text = clean_text(text)
    text = remove_punctuation(text)
    text = to_lowercase(text)
    text = remove_stopwords(text)
    text = lemmatize_text(text)
    return text
```

```python
# Apply preprocessing
print("Applying preprocessing...")
df['clean_text'] = df['text'].apply(preprocess_text_pipeline)
print(f"Total of {len(df)} documents processed")
```

```python
# Function to save labels and IDs for later use
def save_labels_and_ids(df, output_file='emotionverse_labels.csv'):
    labels_df = df[['sentiment', 'primary_emotion']].copy()
    labels_df['id'] = list(range(len(df)))  # Add ID column
    labels_df.to_csv(output_file, index=False)
    print(f"Labels saved to {output_file}")
    return labels_df
```

```python
# 1. Bag of Words (BoW)
def extract_and_save_bow(df, max_features=5000, ngram_range=(1, 1), output_file='bow_features.npz'):
    print("Extracting Bag of Words features...")
    vectorizer = CountVectorizer(max_features=max_features, ngram_range=ngram_range)
    X_bow = vectorizer.fit_transform(df['clean_text'])
    
    # Save sparse matrix
    sp.save_npz(output_file, X_bow)
    
    # Save feature names and vocabulary for reference
    feature_names = vectorizer.get_feature_names_out()
    vocab = vectorizer.vocabulary_
    with open(output_file.replace('.npz', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names.tolist(),
            'vocab_size': len(feature_names),
            'shape': X_bow.shape,
            'ngram_range': ngram_range,
            'max_features': max_features
        }, f)

    print(f"BoW Dimensions: {X_bow.shape}")
    print(f"Feature space size: {len(feature_names)}")
    print(f"Saved to {output_file} and {output_file.replace('.npz', '_metadata.json')}")
    return X_bow, vectorizer
```

```python
# 2. TF-IDF
def extract_and_save_tfidf(df, max_features=5000, ngram_range=(1, 1), output_file='tfidf_features.npz'):
    print("Extracting TF-IDF features...")
    vectorizer = TfidfVectorizer(max_features=max_features, ngram_range=ngram_range)
    X_tfidf = vectorizer.fit_transform(df['clean_text'])
    
    # Save sparse matrix
    sp.save_npz(output_file, X_tfidf)
    
    # Save feature names and vocabulary for reference
    feature_names = vectorizer.get_feature_names_out()
    vocab = vectorizer.vocabulary_
    with open(output_file.replace('.npz', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names.tolist(),
            'vocab_size': len(feature_names),
            'shape': X_tfidf.shape,
            'ngram_range': ngram_range,
            'max_features': max_features
        }, f)
    
    print(f"TF-IDF Dimensions: {X_tfidf.shape}")
    print(f"Feature space size: {len(feature_names)}")
    print(f"Saved to {output_file} and {output_file.replace('.npz', '_metadata.json')}")
    return X_tfidf, vectorizer
```

```python
# 3. Sentiment Lexicon Aligned with Vocabulary
def extract_aligned_lexicon_features(df, vectorizer, output_file='aligned_lexicon_features.npz'):
    print("Extracting Sentiment Lexicon Aligned features...")
    
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    
    # Get vocabulary from the vectorizer
    feature_names = vectorizer.get_feature_names_out()
    
    # Create lexicon scores for each word in the vocabulary
    lexicon_scores = np.zeros((len(feature_names), 4))  # neg, neu, pos, compound
    
    for i, word in tqdm(enumerate(feature_names), total=len(feature_names), desc="Calculating sentiment scores"):
        scores = sid.polarity_scores(word)
        lexicon_scores[i, 0] = scores['neg']
        lexicon_scores[i, 1] = scores['neu']
        lexicon_scores[i, 2] = scores['pos']
        lexicon_scores[i, 3] = scores['compound']
    
    # Save as numpy array
    np.save(output_file, lexicon_scores)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names.tolist(),
            'score_types': ['negative', 'neutral', 'positive', 'compound'],
            'shape': lexicon_scores.shape,
            'lexicon': 'VADER'
        }, f)
    
    print(f"Aligned Lexicon Feature Dimensions: {lexicon_scores.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    # Create and save document-level features based on scores
    document_lexicon_features = np.zeros((len(df), 4))
    X = vectorizer.transform(df['clean_text'])
    
    # For each document, calculate the weighted average of sentiment scores
    for i in tqdm(range(len(df)), desc="Creating document features"):
        # Get indices of non-zero words and their TF-IDF/BoW values
        row = X[i].toarray().flatten()
        nonzero_indices = np.nonzero(row)[0]
        weights = row[nonzero_indices]
        
        # Skip if no terms are found
        if len(nonzero_indices) == 0:
            continue
            
        # Compute weighted average of scores for each sentiment dimension
        weighted_scores = np.zeros(4)
        for j, idx in enumerate(nonzero_indices):
            weighted_scores += weights[j] * lexicon_scores[idx]
            
        # Normalize by the sum of weights
        document_lexicon_features[i] = weighted_scores / weights.sum()
    
    # Save document-level features
    doc_output_file = output_file.replace('.npy', '_document.npy')
    np.save(doc_output_file, document_lexicon_features)
    
    # Save document-level metadata
    with open(doc_output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'score_types': ['negative', 'neutral', 'positive', 'compound'],
            'shape': document_lexicon_features.shape,
            'lexicon': 'VADER',
            'description': 'Document-level sentiment scores weighted by term frequency'
        }, f)
    
    print(f"Document-level Lexicon Feature Dimensions: {document_lexicon_features.shape}")
    print(f"Saved to {doc_output_file}")
    
    return lexicon_scores, document_lexicon_features
```

```python

#EXTRAVADER
def extract_vader_lexicon_features(df, vectorizer, text_column='clean_text', output_file='vader_lexicon_features.npy'):
    """
    Extract VADER sentiment lexicon features aligned with vocabulary.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the text data
    vectorizer : sklearn vectorizer
        Fitted vectorizer (TfidfVectorizer or CountVectorizer)
    text_column : str
        Column name containing the processed text
    output_file : str
        Path to save the numpy array with features
        
    Returns:
    --------
    tuple
        (word_level_features, document_level_features)
    """
    print("Extracting VADER Sentiment Lexicon features...")
    
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    
    # Get vocabulary from the vectorizer
    feature_names = vectorizer.get_feature_names_out()
    
    # Create lexicon scores for each word in the vocabulary
    lexicon_scores = np.zeros((len(feature_names), 4))  # neg, neu, pos, compound
    
    for i, word in tqdm(enumerate(feature_names), total=len(feature_names), desc="Calculating VADER scores"):
        scores = sid.polarity_scores(word)
        lexicon_scores[i, 0] = scores['neg']
        lexicon_scores[i, 1] = scores['neu']
        lexicon_scores[i, 2] = scores['pos']
        lexicon_scores[i, 3] = scores['compound']
    
    # Save as numpy array
    np.save(output_file, lexicon_scores)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names.tolist(),
            'score_types': ['negative', 'neutral', 'positive', 'compound'],
            'shape': lexicon_scores.shape,
            'lexicon': 'VADER'
        }, f)
    
    print(f"Word-level Lexicon Feature Dimensions: {lexicon_scores.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    # Create and save document-level features based on scores
    document_lexicon_features = np.zeros((len(df), 4))
    X = vectorizer.transform(df[text_column])
    
    # For each document, calculate the weighted average of sentiment scores
    for i in tqdm(range(len(df)), desc="Creating document features"):
        # Get indices of non-zero words and their TF-IDF/BoW values
        row = X[i].toarray().flatten()
        nonzero_indices = np.nonzero(row)[0]
        weights = row[nonzero_indices]
        
        # Skip if no terms are found
        if len(nonzero_indices) == 0:
            continue
            
        # Compute weighted average of scores for each sentiment dimension
        weighted_scores = np.zeros(4)
        for j, idx in enumerate(nonzero_indices):
            weighted_scores += weights[j] * lexicon_scores[idx]
            
        # Normalize by the sum of weights
        document_lexicon_features[i] = weighted_scores / weights.sum()
    
    # Save document-level features
    doc_output_file = output_file.replace('.npy', '_document.npy')
    np.save(doc_output_file, document_lexicon_features)
    
    # Save document-level metadata
    with open(doc_output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'score_types': ['negative', 'neutral', 'positive', 'compound'],
            'shape': document_lexicon_features.shape,
            'lexicon': 'VADER',
            'description': 'Document-level sentiment scores weighted by term frequency'
        }, f)
    
    print(f"Document-level Lexicon Feature Dimensions: {document_lexicon_features.shape}")
    print(f"Saved to {doc_output_file}")
    
    return lexicon_scores, document_lexicon_features
```

```python
# EXTRA NRC

def extract_nrc_lexicon_features(df, vectorizer, nrc_lexicon_path, text_column='clean_text', output_file='nrc_lexicon_features.npy'):
    """
    Extract NRC sentiment lexicon features aligned with vocabulary.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the text data
    vectorizer : sklearn vectorizer
        Fitted vectorizer (TfidfVectorizer or CountVectorizer)
    nrc_lexicon_path : str
        Path to the NRC lexicon file
    text_column : str
        Column name containing the processed text
    output_file : str
        Path to save the numpy array with features
        
    Returns:
    --------
    tuple
        (word_level_features, document_level_features)
    """
    print("Extracting NRC Sentiment Lexicon features...")
    
    # Emotions and sentiment categories in NRC
    nrc_emotions = ['anger', 'fear', 'anticipation', 'trust', 'surprise',
                   'sadness', 'joy', 'disgust', 'positive', 'negative']
    
    # Load NRC Lexicon
    nrc = {}
    with open(nrc_lexicon_path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split('\t')
            if len(parts) != 3:
                continue  # Skip malformed lines
                
            word, emotion, association = parts
            if word not in nrc:
                nrc[word] = {}
            nrc[word][emotion] = int(association)
    
    # Get vocabulary from the vectorizer
    feature_names = vectorizer.get_feature_names_out()
    
    # Create lexicon scores for each word in the vocabulary
    lexicon_scores = np.zeros((len(feature_names), len(nrc_emotions)))  # 10 emotion categories
    
    for i, word in tqdm(enumerate(feature_names), total=len(feature_names), desc="Calculating NRC scores"):
        for j, emotion in enumerate(nrc_emotions):
            lexicon_scores[i, j] = nrc.get(word, {}).get(emotion, 0)
    
    # Save as numpy array
    np.save(output_file, lexicon_scores)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names.tolist(),
            'score_types': nrc_emotions,
            'shape': lexicon_scores.shape,
            'lexicon': 'NRC'
        }, f)
    
    print(f"Word-level Lexicon Feature Dimensions: {lexicon_scores.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    # Create and save document-level features based on scores
    document_lexicon_features = np.zeros((len(df), len(nrc_emotions)))
    X = vectorizer.transform(df[text_column])
    
    # For each document, calculate the weighted average of sentiment scores
    for i in tqdm(range(len(df)), desc="Creating document features"):
        # Get indices of non-zero words and their TF-IDF/BoW values
        row = X[i].toarray().flatten()
        nonzero_indices = np.nonzero(row)[0]
        weights = row[nonzero_indices]
        
        # Skip if no terms are found
        if len(nonzero_indices) == 0:
            continue
            
        # Compute weighted average of scores for each dimension
        weighted_scores = np.zeros(len(nrc_emotions))
        for j, idx in enumerate(nonzero_indices):
            weighted_scores += weights[j] * lexicon_scores[idx]
            
        # Normalize by the sum of weights
        document_lexicon_features[i] = weighted_scores / weights.sum()
    
    # Save document-level features
    doc_output_file = output_file.replace('.npy', '_document.npy')
    np.save(doc_output_file, document_lexicon_features)
    
    # Save document-level metadata
    with open(doc_output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'score_types': nrc_emotions,
            'shape': document_lexicon_features.shape,
            'lexicon': 'NRC',
            'description': 'Document-level NRC emotion scores weighted by term frequency'
        }, f)
    
    print(f"Document-level Lexicon Feature Dimensions: {document_lexicon_features.shape}")
    print(f"Saved to {doc_output_file}")
    
    return lexicon_scores, document_lexicon_features
```

```python
# COMBINE FEATURES :

def combine_features(tfidf_features, bow_features, vader_features=None, nrc_features=None, embedding_features=None):
    """
    Combine different feature sets for later use in classification.
    
    Parameters:
    -----------
    tfidf_features : scipy.sparse matrix
        TF-IDF features
    bow_features : scipy.sparse matrix
        Bag of Words features
    vader_features : numpy.ndarray, optional
        VADER sentiment features
    nrc_features : numpy.ndarray, optional
        NRC emotion features
    embedding_features : numpy.ndarray, optional
        Text embedding features
        
    Returns:
    --------
    dict
        Dictionary with feature matrices and metadata
    """
    from scipy.sparse import hstack, csr_matrix
    import scipy.sparse as sp
    
    features = {
        'tfidf': tfidf_features,
        'bow': bow_features
    }
    
    if vader_features is not None:
        features['vader'] = csr_matrix(vader_features)
        
    if nrc_features is not None:
        features['nrc'] = csr_matrix(nrc_features)
        
    if embedding_features is not None:
        features['embeddings'] = csr_matrix(embedding_features)
    
    # Create a combined sparse feature matrix
    feature_matrices = [features[key] for key in features]
    combined_features = hstack(feature_matrices)
    
    features['combined'] = combined_features
    
    # Create feature information
    feature_info = {
        'shapes': {k: v.shape for k, v in features.items()},
        'dtypes': {k: str(v.dtype) for k, v in features.items()},
        'sparsity': {k: 1.0 - (v.nnz / (v.shape[0] * v.shape[1])) 
                    if sp.issparse(v) else np.count_nonzero(v) / v.size
                    for k, v in features.items()}
    }
    
    features['info'] = feature_info
    
    return features
```

```python
# 4. BERT Embeddings
def extract_and_save_bert_embeddings(df, output_file='bert_features.npy', model_name='bert-base-uncased', pooling_method='mean', batch_size=8):
    print(f"Extracting BERT embeddings using {model_name}...")
    
    # Load BERT model and tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    
    # Move model to GPU if available
    model = model.to(device)
    model.eval()
    
    # Function to extract embeddings in batch
    def get_bert_embeddings_batch(texts, max_length=512):
        # Tokenize and prepare for BERT
        inputs = tokenizer(texts, return_tensors='pt', max_length=max_length, 
                          padding='max_length', truncation=True)
        inputs = {k: v.to(device) for k, v in inputs.items()}
        
        # Get embeddings
        with torch.no_grad():
            outputs = model(**inputs)
        
        # Different pooling strategies
        if pooling_method == 'cls':
            # Use [CLS] token (first token)
            embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
        elif pooling_method == 'mean':
            # Mean of all token embeddings (excluding padding)
            attention_mask = inputs['attention_mask']
            masked_embeddings = outputs.last_hidden_state * attention_mask.unsqueeze(-1)
            sum_embeddings = masked_embeddings.sum(1)
            sum_mask = attention_mask.sum(1, keepdim=True)
            embeddings = (sum_embeddings / sum_mask).cpu().numpy()
        
        return embeddings
    
    # Extract embeddings in batches
    embedding_dim = model.config.hidden_size
    X_bert = np.zeros((len(df), embedding_dim))
    
    # Process in batches
    for i in tqdm(range(0, len(df), batch_size), desc="Creating BERT embeddings"):
        batch_texts = df['clean_text'].iloc[i:i+batch_size].tolist()
        batch_embeddings = get_bert_embeddings_batch(batch_texts)
        
        # Store embeddings
        end_idx = min(i + batch_size, len(df))
        X_bert[i:end_idx] = batch_embeddings
    
    # Save as numpy array
    np.save(output_file, X_bert)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'model': model_name,
            'embedding_dim': embedding_dim,
            'pooling_method': pooling_method,
            'shape': X_bert.shape
        }, f)
    
    print(f"BERT Embedding Dimensions: {X_bert.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    return X_bert
```

```python
# 5. Sentiment-Specific BERT
def extract_and_save_sentiment_bert(df, output_file='sentiment_bert_features.npy', 
                                  model_name='nlptown/bert-base-multilingual-uncased-sentiment'):
    print(f"Extracting sentiment features using {model_name}...")
    
    # Load pre-trained sentiment analysis model
    sentiment_analyzer = pipeline('sentiment-analysis', model=model_name, device=0 if torch.cuda.is_available() else -1)
    
    # Extract sentiment scores
    sentiment_scores = []
    
    # Process in batches to avoid memory issues
    batch_size = 16
    
    for i in tqdm(range(0, len(df), batch_size), desc="Extracting sentiment scores"):
        end_idx = min(i + batch_size, len(df))
        batch_texts = df['text'].iloc[i:end_idx].tolist()
        
        try:
            batch_results = sentiment_analyzer(batch_texts)
            sentiment_scores.extend(batch_results)
        except Exception as e:
            print(f"Error processing batch {i}-{end_idx}: {e}")
            # In case of error, append default results
            for _ in range(end_idx - i):
                sentiment_scores.append({'label': 'UNKNOWN', 'score': 0.0})
    
    # Extract labels and scores
    labels = [result['label'] for result in sentiment_scores]
    scores = [result['score'] for result in sentiment_scores]
    
    # Convert labels to numeric values (if needed)
    # Example for model nlptown/bert-base-multilingual-uncased-sentiment:
    # Labels are 1 to 5 stars as strings ('1', '2', ..., '5')
    numeric_labels = []
    for label in labels:
        try:
            if isinstance(label, str) and label.isdigit():
                numeric_labels.append(int(label))
            else:
                # For other label formats
                numeric_labels.append(0)  # default value
        except:
            numeric_labels.append(0)
    
    # Create array with scores and labels
    X_sentiment = np.column_stack((numeric_labels, scores))
    
    # Save as numpy array
    np.save(output_file, X_sentiment)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'model': model_name,
            'features': ['sentiment_label', 'confidence_score'],
            'shape': X_sentiment.shape
        }, f)
    
    print(f"Sentiment BERT Feature Dimensions: {X_sentiment.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    return X_sentiment
```

```python
# 6. Sentence-BERT for Semantic Sentence Embeddings
def extract_and_save_sentence_embeddings(df, output_file='sentence_embeddings.npy', 
                                       model_name='all-MiniLM-L6-v2'):
    print(f"Extracting sentence embeddings using {model_name}...")
    
    # Load Sentence-BERT model
    model = SentenceTransformer(model_name)
    model.to(device)
    
    # Extract embeddings for each text
    X_sentence = model.encode(
        df['text'].tolist(),
        batch_size=32,
        show_progress_bar=True,
        device=device,
        convert_to_numpy=True
    )
    
    # Save as numpy array
    np.save(output_file, X_sentence)
    
    # Save metadata
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'model': model_name,
            'embedding_dim': X_sentence.shape[1],
            'shape': X_sentence.shape
        }, f)
    
    print(f"Sentence Embedding Dimensions: {X_sentence.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    return X_sentence
```

```python
# 7. Additional Linguistic Features
def extract_linguistic_features(df, output_file='linguistic_features.npy'):
    print("Extracting additional linguistic features...")
    
    # Download necessary NLTK resources
    nltk.download('punkt')
    
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()
    
    # Lists of intensifier and diminisher words
    intensifiers = [
        'very', 'extremely', 'incredibly', 'absolutely', 'completely', 'totally',
        'utterly', 'highly', 'especially', 'particularly', 'really', 'so', 'too'
    ]
    
    diminishers = [
        'slightly', 'somewhat', 'a bit', 'a little', 'kind of', 'kinda',
        'sort of', 'hardly', 'barely', 'scarcely', 'only', 'just'
    ]
    
    # Negation words
    negations = [
        'not', 'no', 'never', "don't", "doesn't", "didn't", "won't", "wouldn't",
        "isn't", "aren't", "wasn't", "weren't", "haven't", "hasn't", "hadn't",
        "can't", "cannot", "couldn't"
    ]
    
    # Feature set for each document
    features = np.zeros((len(df), 10))
    
    for i, text in tqdm(enumerate(df['text']), total=len(df), desc="Extracting linguistic features"):
        # Tokenize
        tokens = word_tokenize(text.lower())
        
        # 1. Text length (in tokens)
        features[i, 0] = len(tokens)
        
        # 2. Average word length
        if tokens:
            features[i, 1] = np.mean([len(token) for token in tokens if token.isalpha()])
        
        # 3. Count of intensifier words
        features[i, 2] = sum(1 for token in tokens if token in intensifiers)
        
        # 4. Count of diminisher words
        features[i, 3] = sum(1 for token in tokens if token in diminishers)
        
        # 5. Count of negations
        features[i, 4] = sum(1 for token in tokens if token in negations)
        
        # 6–9. VADER sentiment scores (neg, neu, pos, compound)
        sentiment_scores = sid.polarity_scores(text)
        features[i, 5] = sentiment_scores['neg']
        features[i, 6] = sentiment_scores['neu']
        features[i, 7] = sentiment_scores['pos']
        features[i, 8] = sentiment_scores['compound']
        
        # 10. Positive/Negative ratio
        if sentiment_scores['neg'] > 0:
            features[i, 9] = sentiment_scores['pos'] / sentiment_scores['neg']
        else:
            features[i, 9] = sentiment_scores['pos'] * 2  # Arbitrary value when there's no negativity
    
    # Save as numpy array
    np.save(output_file, features)
    
    # Save metadata
    feature_names = [
        'text_length', 'avg_word_length', 'intensifier_count', 'diminisher_count',
        'negation_count', 'vader_neg', 'vader_neu', 'vader_pos', 'vader_compound',
        'pos_neg_ratio'
    ]
    
    with open(output_file.replace('.npy', '_metadata.json'), 'w') as f:
        json.dump({
            'feature_names': feature_names,
            'shape': features.shape
        }, f)
    
    print(f"Linguistic Feature Dimensions: {features.shape}")
    print(f"Saved to {output_file} and {output_file.replace('.npy', '_metadata.json')}")
    
    return features
```

```python
# Main function to extract all features
def extract_all_advanced_features(df, output_dir='emotionverse_features_advanced'):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Define file paths
    bow_file = os.path.join(output_dir, 'bow_features.npz')
    tfidf_file = os.path.join(output_dir, 'tfidf_features.npz')
    aligned_lexicon_file = os.path.join(output_dir, 'aligned_lexicon_features.npy')
    bert_file = os.path.join(output_dir, 'bert_features.npy')
    sentiment_bert_file = os.path.join(output_dir, 'sentiment_bert_features.npy')
    sentence_emb_file = os.path.join(output_dir, 'sentence_embeddings.npy')
    linguistic_file = os.path.join(output_dir, 'linguistic_features.npy')
    labels_file = os.path.join(output_dir, 'emotionverse_labels.csv')
    extravader = os.path.join(output_dir, 'extravader.npy')
    extranrc =  os.path.join(output_dir, 'extranrc.npy')
    combined =  os.path.join(output_dir, 'combined.npy')
    
    # Extract and save features
    print("\n--- Starting advanced feature extraction ---\n")
    
    # 1. Extract BoW
    X_bow, bow_vectorizer = extract_and_save_bow(df, output_file=bow_file)
    
    # 2. Extract TF-IDF
    X_tfidf, tfidf_vectorizer = extract_and_save_tfidf(df, output_file=tfidf_file)
    
    # 3. Extract aligned sentiment lexicon (using TF-IDF for weighting)
    X_lexicon, X_doc_lexicon = extract_aligned_lexicon_features(df, tfidf_vectorizer, output_file=aligned_lexicon_file)
    
    _, vader_features = extract_vader_lexicon_features(df, tfidf_vectorizer,output_file=extravader )
    _, nrc_features = extract_nrc_lexicon_features(df, tfidf_vectorizer, output_file=extranrc)
    
    
    scaler_vader = StandardScaler()
    scaled_vader_features = scaler_vader.fit_transform(vader_features)

    scaler_nrc = StandardScaler()
    scaled_nrc_features = scaler_nrc.fit_transform(nrc_features)
    
    from sklearn.preprocessing import MaxAbsScaler
    sparse_scaler = MaxAbsScaler()
    scaled_tfidf = sparse_scaler.fit_transform(X_tfidf)
    scaled_bow = sparse_scaler.fit_transform(X_bow)
    
    all_features = combine_features(
    scaled_tfidf, 
    scaled_bow, 
    scaled_vader_features, 
    scaled_nrc_features
    )

    # 4. Extract additional linguistic features
    X_linguistic = extract_linguistic_features(df, output_file=linguistic_file)
    
    # 5. Extract BERT embeddings
    X_bert = extract_and_save_bert_embeddings(df, output_file=bert_file)
    
    # 6. Extract sentiment features using BERT
    try:
        X_sentiment_bert = extract_and_save_sentiment_bert(df, output_file=sentiment_bert_file)
    except Exception as e:
        print(f"Error extracting sentiment BERT features: {e}")
        print("Skipping this extraction...")
    
    # 7. Extract sentence embeddings with Sentence-BERT
    X_sentence = extract_and_save_sentence_embeddings(df, output_file=sentence_emb_file)
    
    # 8. Save labels
    labels_df = save_labels_and_ids(df, output_file=labels_file)
    
    print("\n--- Feature extraction completed! ---")
    print(f"All features saved in {output_dir}/")
    
    return {
        'bow': (X_bow, bow_vectorizer),
        'tfidf': (X_tfidf, tfidf_vectorizer),
        'aligned_lexicon': (X_lexicon, X_doc_lexicon),
        'bert': X_bert,
        'sentence_embeddings': X_sentence,
        'linguistic': X_linguistic,
        'labels': labels_df
        'scaledtfidf' : scaled_tfidf
        'scaledbow' : scaled_bow
        'extravader' : scaled_vader_features
        'extranrc' : scaled_nrc_features
        'Combined' : all_features
    }
```
